 ## 関数

 関数の定義は`def`ではじまる．

```python
def square(x):
    return x * x
```

```python
square
```

 定義した関数`square`に実引数`2`を与えて呼び出す．

```python
square(2)
```

 関数を引数として受け取る関数も定義できる．

```python
def apply_twice(f, x):
    return f (f(x))
```

```python
apply_twice(square, 2)
```

 C言語と異なり，関数内部で新しい関数を定義してそれ（内部で定義した関数自体）を返すこともできる．

```python
def twice(f):
    def ff(x):
        return f(f(x))
    return ff    
```

```python
square2 = twice(square)
square2(2)
```
